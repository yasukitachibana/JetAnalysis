```python
import numpy as np

lim = [-0.05,0.35]
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection='3d')



file ='/Users/yasukitachibana/GoogleDrive/Codes/JetAnalysis/JetShape2D/5020_PP_New/'
file = file + 'scaled_2d_hist_total_JetShape2D_r_jetr0.4_ptj120-300_rapj0.0-1.6_pt0.7-300.0_rap0.0-2.4_mixed_ev.txt'


data = np.loadtxt(file, comments='#')
#data2 = np.loadtxt(file2, comments='#')

ax.set_xlabel(r'$\Delta\phi$', fontsize=13, labelpad=14)
ax.set_ylabel(r'$\Delta\eta$', fontsize=13, labelpad=14)



x = data[:,1]
dx = data[:,2]-data[:,1]
y = data[:,4]
dy = data[:,5]-data[:,4]
dz = data[:,6]
z = np.zeros_like(dz)

cmap = cm.get_cmap('jet') # Get desired colormap - you can change this!
max_height = np.max(dz)   # get range of colorbars so we can normalize
min_height = np.min(dz)
# scale each z to [0,1], and get their rgb values

rgba = [cmap((k-lim[0])/lim[1]) for k in dz] 
# rgba = [cmap((k-lim[0])/lim[1]) for k in dz] 

# ax.set_zlim(lim[0],lim[1])
ax.view_init(elev=20, azim=150)
max_height = np.max(dz)
min_height = np.min(dz)
# scale each z to [0,1], and get their rgb values
rgba = [cmap((k-min_height)/max_height) for k in dz] 

ax.bar3d(x, y, z, dx, dy, dz, color=rgba)
fig.savefig("pp_full_rap1-2_test.pdf")



```
